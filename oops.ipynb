```python
# OOPS- FUNCTION
#EVEN ODD
def even_odd(a,b):
    """here a start--->start number
    b--->end number"""
    if a>b:
        a,b=b,a
        en=[x for x in range(a,b) if x%2==0]
        od=[x for x in range(a,b) if x%2!=0]
        print('even nos are: {}\nodd nos are:{}'.format(en,od))
```

```python
even_odd(40,3)
```

```python
# LEAP YEAR
def leap_year(a,b):
    if a>b:
        a,b=b,a
    ly=[x for x in range(a,b) if (x%4==0 and x%100!=0)or (x%400==0)]
    print('no of leap years are:{}\n leap years are:{}'.format(len(ly),ly))
```

```python
leap_year(2023,2002)
```

```python
# TABLES
def table(a,sv):
    i=1
    t1=[]
    while i<=sv:
        t1.append(i*a)
        i+=1
    print(t1)
```

```python
table(12,10)
table(12,5)
table(17,15)
```

```python
# WHOLE NUMBERS AND DECIMAL NUMBERS
def twd(*args):
    text=[x for x in args if type(x)==str]
    wn=[x for x in args if type(x)==int]
    dn=[x for x in args if type(x)==float]
    print('in given mix text are:{} \nwn nos are:{} \ndn nos are:{}'.format(text,wn,dn))
```

```python
# FUTURE VALUE
#FV=PV*(1+r)^n
plan={'amt':[10000,20000,30000],
      'rate':[0.07,0.08,0.06],
      'year':[3,5,7,10]}
```

```python
def fv(**kwargs):
    result={'pv':[],'ir':[],'yr':[],'fv':[]}
    for pv in kwargs['amt']:
        for r in kwargs['rate']:
            for n in kwargs['year']:
                f=round(pv*(1+r)**n,0)
                result['pv'].append(pv)
                result['ir'].append(r)
                result['yr'].append(n)
                result['fv'].append(f)
    import pandas
    data=pandas.DataFrame(result)
    print(data)
```

```python
fv(**plan)
```
